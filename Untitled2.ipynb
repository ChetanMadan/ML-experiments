```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
oliv = datasets.fetch_olivetti_faces()
```

```python
oliv.keys()
```

```python
oliv.data.shape
```

```python
fig =plt.figure(figsize=(20,20))
for i in range(100):
    ax = fig.add_subplot(10,10,i+1)
    ax.imshow(oliv.images[i], cmap=plt.cm.bone)
plt.show()
```

```python
x,y = oliv.data, oliv.target
pca_oliv = PCA()
pca_oliv.fit(x)
```

```python
k = 0
total = 0
while total < 0.95:
    total = total + pca_oliv.explained_variance_ratio_[k]
    k = k+1
k
```

```python
pca_oliv = PCA(n_components = k)
x_transformed = pca_oliv.fit_transform(x)
x_transformed.shape
```

```python
x_approx = pca_oliv.inverse_transform(x_transformed)
x_approx.shape
```

```python
x_approx = x_approx.reshape((400,64,64))
```

```python
fig =plt.figure(figsize=(20,20))
for i in range(100):
    ax = fig.add_subplot(10,10,i+1)
    ax.imshow(x_approx[i], cmap=plt.cm.bone)
plt.show()
```

```python
pca_oliv.components_.shape
```

```python
eigen_v = pca_oliv.components_.reshape(123, 64, 64)
```

```python
fig =plt.figure(figsize=(20,20))
for i in range(100):
    ax = fig.add_subplot(10,10,i+1)
    ax.imshow(eigen_v[i], cmap=plt.cm.bone)
plt.show()
```
