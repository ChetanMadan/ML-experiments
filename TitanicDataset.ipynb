```python
#fill empty age with mean 

titanic.Age.fillna(titanic.Age.mean(), inplace=True)
```

```python
#remove empty embarked

titanic.Embarked.dropna(inplace=True)
```

```python
def getNumber(s):
    if (s == 'male'):
        return 1
    else:
        return 2

titanic['gender'] = titanic['Sex'].apply(getNumber)
del titanic['Sex']


#.apply applies a particular function to a particular column
```
