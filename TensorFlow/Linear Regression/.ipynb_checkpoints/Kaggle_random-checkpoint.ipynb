```python
#load dataset
import pandas as pd

train = pd.read_csv("./Kaggle_random/train.csv")
test = pd.read_csv("./Kaggle_random/test.csv")

train_x = train.x.values
train_y = train.y.values
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (16, 10)

plt.scatter(train_x, train_y)
plt.show()
```

```python
import tensorflow as tf
```

```python
m0 = tf.Variable(1.)
m1 = tf.Variable(1.)

epochs = 1000

y = m1*train_x + m0

loss = tf.reduce_mean(tf.square(y-train_y))

print(tf.Session().run(loss))

optimizer = tf.train.GradientDescentOptimizer(0.1)
train = optimizer.minimize(loss)
init = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(init)
    
    train_data = []
    for steps in range(epochs):
        evals = sess.run([train,m0,m1])[1:]
        if steps%50==0:
            print ('step {} of {} completed, current values = {}'.format(steps, epochs, evals))
            train_data.append(evals)
```
