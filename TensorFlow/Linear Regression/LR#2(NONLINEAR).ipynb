```python
#generate datapoints

import numpy as np

x_data = np.random.rand(100).astype(np.float32)
y_data = x_data**2

y_data = np.vectorize(lambda y: y + np.random.normal(loc=0.0, scale=0.03))(y_data)
```

```python
#plotting these points

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (16, 10)

plt.scatter(x_data, y_data)
plt.show()
```

```python
#expected line
#m0 = 0, m1=0, m2=1

xs = np.arange(0,1,0.01)
ys = xs**2

plt.scatter(x_data, y_data)
plt.plot(xs,ys,'r')
plt.show()
```

```python
#ml madness

import tensorflow as tf
```

```python
m0 = tf.Variable(1.)
m1 = tf.Variable(1.)
m2 = tf.Variable(1.)

epochs = 50000

y = m2*(x_data**2) + m1*x_data + m0

loss = tf.reduce_mean(tf.square(y-y_data))

optimizer = tf.train.GradientDescentOptimizer(0.01)
train = optimizer.minimize(loss)
init = tf.global_variables_initializer()    

with tf.Session() as sess:
    sess.run(init)
    
    train_data = []
    
    for steps in range(epochs):
        evals = sess.run([train,m0,m1,m2])[1:]
        
        if steps % 1000 == 0:
            print('steps {} of {} completed, current value = {}'.format(steps, epochs, evals))
            train_data.append(evals)
```

```python
#expected line
#m0 = 0, m1=0, m2=1

xs = np.arange(0,1,0.01)
ys = xs**2
plt.plot(xs,ys,'r', label='actual')


#predicted line
xs2 = np.arange(0,1,0.01) 
ys2 = train_data[-1][2]*np.square(xs2) + train_data[-1][1]*xs2 + train_data[-1][0]
plt.plot(xs2,ys2, label='predicted')

plt.legend()
plt.show()
```
